```python
import pandas as pd
```

```python
#Crear un DataFrame de 1000000 de publicaciones de instagram 
cantidad = 1000000
```

```python
import random
```

```python
#Datos aleatorios
influencer = ["Ibai Llanos", "Kimberly Loaiza", "Luisito Comunica", "AuronPlay", "Yuya", "El Rubius", "Rosalía", "MrBeast"]
likes = [125400, 890000, 312000, 450000, 180000, 210500, 940000, 3200000]
comentarios = ["wow", "Fantastico", "Puede mejorar", "Terrible", "no me gustó", "Fake", "ward", "hay motivación, falta talento"]
fecha = ["2026-05-01", "2026-05-02", "2026-05-03", "2026-05-04", "2026-05-05", "2026-05-06", "2026-05-07", "2026-05-07"]
tipo_post = ["Reel", "Video Musical", "Foto de Viaje", "Meme/Imagen", "Carrusel", "Gameplay Short", "Fashion Editorial", "Giveaway Video"]
```

```python
#Creación de los DataFrames
dfMillon = pd.DataFrame({
    "influencer" : [random.choice(influencer) for _ in range(cantidad)],
    "likes" : [random.choice(likes) for _ in range(cantidad)],
    "comentarios" : [random.choice(comentarios) for _ in range(cantidad)],
    "fecha" : [random.choice(fecha) for _ in range(cantidad)],
    "tipo_post" : [random.choice(tipo_post) for _ in range(cantidad)]
})
```
