```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import ( accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay )
```

```python
## Cargamos el dataSet Iris
iris = load_iris (as_frame = True)
```

```python
print (iris.DESCR)
```

```python
dfIris = iris.frame
```

```python
dfIris.head()
```

## Exploración del conjunto de datos

```python
dfIris.shape
```

```python
dfIris.columns
```

```python
iris.target_names
```

## Renombrar la columna target

```python
dfIris.rename(columns = {"target" : "especies"}, inplace = True)
```

```python
dfIris["especies"] = dfIris["especies"].map({
    0 : "setosa",
    1 : "versicolor",
    2 : "viginica"
})
```

```python
dfIris.head()
```

```python
dfIris.tail()
```

## Identificar variables predictoras y variable objetivo

```python
# Variable predictora 
X = dfIris.drop(columns = ["especies"])
```

```python
print ("Variables predictoras")
print (X.columns)
```

```python
# Variable objetivo 
y = dfIris["especies"]
```

```python
print ("Variable objetivo")
print (y.name)
```

## Dividir los datos en prueba y entrenamiento 

Distribuir los datos en 80% para entrenamiento y 20% para prueba

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.20,
    random_state = 42,
    stratify = y
)
```

```python
print ("X_train", X_train.shape)
print ("X_test", X_test.shape)
print ("y_train", y_train.shape)
print ("y_test", y_test.shape)
```

## Escalado de variable

```python
scaler = StandardScaler()
```

```python
# Escalamos las variables
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
print ("Sin escalar")
print (X_train.head())
print ("Escalado")
print (X_train_scaled[:5])
```

## Entrenamieto del modelo KNN
En esta prueba utilizaremos k = 3

```python
modelo = KNeighborsClassifier(n_neighbors = 3)
```

```python
modelo.fit(
    X_train_scaled, y_train
)
```

  ## Realizar las predicciones con las variables de prueba

```python
prediccion = modelo.predict(
    X_test_scaled
)
```

```python
print (prediccion[:10])
```

```python
# Comparación entre valores reales y predecibles
comparacion = pd.DataFrame({
    "especie_real" : y_test.values,
    "especia_predicha" : prediccion
})
```

```python
print (comparacion.head(10))
```

## Evaluación del modelo 

```python
accuracy = accuracy_score(
    y_test,
    prediccion
)
```

```python
print(f"Accuracy/exactitud : {accuracy: .4f}")
```

```python
## Matríz de confusión 

cm = confusion_matrix(
    y_test,
    prediccion
)
```

```python
cm_df = pd.DataFrame(
    cm, 
    index = modelo.classes_,
    columns = modelo.classes_
)
cm_df
```

## Classification report
Proporciona métricas que permiten evaluar el desempeño de cada clase: precision, recall, f1-score, support

```python
print (classification_report(y_test, prediccion))
```
