## Parte 4. Selección de variables Predictorias y Variable objetivo 

**X** - Variables que utilizará el modelo 

**Y** - Variables que queremos predecir 

```python
columnas = [
    "Pclass",
    "Sex",
    "Age",
    "SibSp",
    "Parch",
    "Fare",
    "Embarked"
]
```

```python
## Definimos x y 
X = dfTitanic[columnas]
y = dfTitanic["Survived"]
```

```python
print(X.columns.tolist())
```

```python
print(y.name)
```

## Parte 5. Tratamiento de los 

Para simplificar el proceso, realizamos la imputación de los datos antes de la separación TRAIN/TEST

```python
# Revisamos las variables predictoras
X.isnull().sum()
```

```python
X["Age"] = X["Age"].fillna(
    X["Age"].median()
)
```

```python
X["Embarked"] = X["Embarked"].fillna(
    X["Embarked"].mode()[0]
)
```

```python
X.isnull().sum()
```

```python
X.head()
```

## Parte 6. Escalado de los datos

Los algoritmos no entienden texto. Se deben convertir las variable categoricas en variables numéricas

```python
X = pd.get_dummies(
    X,
    columns = ["Sex", "Embarked"],
    drop_first = True
)
```

```python
X.head()
```

## Parte 7. Particionamos los datos 

No debemos entrenar y evaluar utilizando los mismos datos. ya que el algoritmos va a memorizar 

La recomendación es: 

Entrenamiento 80%
Pruebas 20%

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.20,
    random_state = 42
)
```

```python
print ("Entrenamiento", X_train.shape)
```

```python
print ("Prueba: ", y_test.shape)
```

## Parte 8. Escalado de las variable 

Las variables numericas tienen valores distintos. El escalado evita que una variable domine más que otra

```python
columnas_numericas = [
    "Age",
    "Fare",
    "SibSp",
    "Parch"
]
```

```python
X_train[columnas_numericas].describe()
```

```python
scaler = StandardScaler()
X_train[columnas_numericas] = scaler.fit_transform(
    X_train[columnas_numericas]
)
X_test[columnas_numericas] = scaler.transform(
    X_test[columnas_numericas]
)
```

```python
X_train[columnas_numericas].describe()
```

## Parte 9. Entrenamiento de los datos 
Usaremos un arbol de decisión para clasificar registros

```python
modelo = DecisionTreeClassifier (
    max_depth = 4,
    random_state = 42
)
```

```python
modelo.fit(
    X_train,
    y_train
)
```

## Parte 10. Evaluamos el modelo

Ralizamos predicciones para evaluar el modelo 

```python
predicciones = modelo.predict(
    X_test
)
```

```python
pd.DataFrame({
    "Real ": y_test,
    "Predicción": predicciones
}).head(25)
```

## Parte 11. Evaluación del modelo

El accuracy determina el procentaje de predicciones correctas realizadas por el modelo 

```python
accuracy = accuracy_score(
    y_test,
    predicciones
)
```

```python
print(f"Accuracy : {accuracy: .4f}")
```

## Parte 12. Importancia de las variables 

Analizar las variables mas relevantes para el algoritmo, al realizar las predicciones.


```python
importancias = pd.DataFrame({
    "Variable" : X_train.columns,
    "Importancia" : modelo.feature_importances_ #Que varibale es importante para el modelo
})
```
